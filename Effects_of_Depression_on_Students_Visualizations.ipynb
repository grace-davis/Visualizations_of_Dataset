```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_csv("effectsofanxiety_depressiononstudents.csv")
```

```python
df.describe()
```

```python
df_1 = df.iloc[:, 12]
df_2 = df.iloc[:, 0]
df_3 = pd.concat([df_1, df_2], axis = 1)
df_ = df_3.rename_axis('Student')
df_ = df_.reset_index('Student')

sns.scatterplot(data=df_, x='Student', y='Mental Health Score', hue = 'Gender')
plt.title('Mental Health Scores (Fig. 1)')
plt.xlabel('Student #')
plt.ylabel('Scores (Range 0-36)')
```

```python
#GRAPHS COMPARING MENTAL HEALTH AMONG GROUPS

#AGE & GENDER
sns.barplot(data = df, x='Age', y = 'Mental Health Score', hue = 'Gender', order=['18 or younger', '19 to 24', '25 or older'])

plt.title('Mental Health Scores by Age & Gender (Fig 2)')
```

```python
#GENDER AND AGE

sns.boxplot(data = df, x='Gender', y = 'Mental Health Score', hue ='Age')
plt.title('Mental Health Scores by Gender & Age (Fig. 3)')
```

```python
#EDUCATION LEVEL AND GENDER
sns.boxplot(data = df, x='Educational Level', y = 'Mental Health Score', hue ='Gender', order=['High School','College - Bachelors','Master'])
plt.title('Mental Health Scores by Education Level & Gender (Fig. 4)')
```
